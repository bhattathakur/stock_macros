### This technical tool is one of the most popular ways for clients to see the market’s top performers. The Relative Strength Rating is the result of calculating a stock’s percentage price change over the last 12 months. A 40% weight is assigned to the latest three-month period; the remaining three quarters each receive 20% weight. All stocks are arranged in order of greatest price percentage change and assigned a percentile rank from 99 (highest) to 1 (lowest).

Found this equation used somewhere to find the RS rating for a stock
    $$ \big(\frac{C-C63}{C63}\times 0.4 + \frac{C-C126}{C126} \times 0.2 + \frac{C-C189}{C189}\times 0.2+\frac{C-C252}{C252}\times 0.2\big) \times 100 $$

# Condition 3: 200 SMA trending up for at least 1 month (ideally 4-5 months)
# Condition 4: 50 SMA> 150 SMA and 50 SMA> 200 SMA
# Condition 5: Current Price > 50 SMA
# Condition 6: Current Price is at least 30% above 52 week low (Many of the best are up 100-300% before coming out of consolidation)
# Condition 7: Current Price is within 25% of 52 week high

```python
def get_sma(df,period):
    """
    gives the sma of a stock for given period
    
    """
    close=df['Close']
    #print(f"closes {closes}")
    
    return close.tail(period).mean().round(2)

def get_200sma_past20(df,period=200):
    """
    gives the 200-sma past 20 days  for given stock
    
    """
    close=df['Close'][:-20]
   
    
    return close.tail(period).mean().round(2)

def get_current_price(df):
    """
    returns the latest close price of a security
    """
    return df['Close'][-1].round(2)

def get_52week_low(ticker):
    """
    returns 52 week low from the given closes
    """
    #return df['Close'].min().round(2)
    return round(ticker.info["fiftyTwoWeekLow"],2)

def get_52week_high(ticker):
    """
    returns 52 week high from the given closes
    """
    #return df['Close'].max().round(2)
    return round(ticker.info["fiftyTwoWeekHigh"],2)


    
def get_rs_value(df,ticker):
    if st in list(df_rs['Ticker']):
        #print("yes")
        #print(f"rs_rank for {ticker}\t {df_rs.loc[df_rs['Ticker']==st]['Percentile'].values[0]}")
        return (df_rs.loc[df_rs['Ticker']==st]['Percentile'].values[0])
    else: 
        #print("no")
        return 0
    
def condition1(df):
    return get_current_price(df)>get_sma(df,150) and get_current_price(df)>get_sma(df,200)

def condition2(df):
    return get_sma(df,150)>get_sma(df,200)

def condition3(df):
    return get_sma(df,200)>get_200sma_past20(df)

def condition4(df):
    return get_sma(df,50)>get_sma(df,150) and get_sma(df,50)>get_sma(df,200)

def condition5(df):
    return get_current_price(df)>get_sma(df,50)

def condition6(df,ticker):
    return get_current_price(df)>=1.30*get_52week_low(ticker)

def condition7(df,ticker):
    return get_current_price(df)>=0.75*get_52week_high(ticker)

def condition8(df,ticker):
    return get_rs_value(df,ticker)>=0



#print(f"get_rs_value {get_rs_value(df_rs, "ABBV")}")  
#condition8(df_rs,ticker

print(get_52week_high(ticker))
```

```python
#%timeit
#s & p 500 companies
#Updated s and p 500 from wikipage

url="https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
dfs=pd.read_html(url)
df=dfs[0]  #first table
df.head()
snp_list=df['Symbol']
#change the BRK.B to BRK-B and BF.B to BF-B
snp_list = ['BRK-B' if item == 'BRK.B' else item for item in snp_list]
snp_list = ['BF-B' if item == 'BF.B' else item for item in snp_list]

snp_list=df['Symbol']#[:5]
mark_criterion=[]
cprice=[]
sma_50=[]
sma_150=[]
sma_200=[]
sma_200_past_20=[]
low_52week=[]
high_52week=[]
RS_rating=[]

#rs values 
df_rs=pd.read_csv('https://raw.githubusercontent.com/skyte/rs-log/main/output/rs_stocks.csv')

j=1
for i in snp_list:
    
    ticker=yf.Ticker(i)
    df=ticker.history(period='52wk',back_adjust=True)
    
    print(f"\n {j}) Checking condition for {i} ...\n")
    j=j+1
     
    #checking all the condition
    try:
        if(condition1(df)):
            if(condition2(df)):
                if(condition3(df)):
                    if(condition4(df)):
                        if(condition5(df)):
                            if(condition6(df,ticker)):
                                if(condition7(df,ticker)):
                                    if(condition8(df_rs,ticker)):
                                        print(20*"--")
                                        print(f"all conditions are satisfied for {i}")
                                        mark_criterion.append(i)
                                        cprice.append(get_current_price(df))
                                        sma_50.append(get_sma(df,50))
                                        sma_150.append(get_sma(df,150))
                                        sma_200.append(get_sma(df,200))
                                        sma_200_past_20.append(get_200sma_past20(df))
                                        low_52week.append(get_52week_low(ticker))
                                        high_52week.append(get_52week_high(ticker))
                                        RS_rating.append(get_rs_value(df_rs,ticker))
                                        print(20*"--")
        else:
            continue
    except:
        print("Error occured !")
        

print("DONE!!!")

#saving in the dataframe
#data frame of the crietiron
from datetime import datetime
fapp=datetime.now().strftime('%m%d%Y')
fname="snp"+fapp+".csv"
print(fname)
df_sp=pd.DataFrame({"S&P-Tickers":mark_criterion,"Current_Price":cprice,"sma_50":sma_50,"sma_150":sma_150,"sma_200":sma_200,"sma200_past20":sma_200_past_20,"RS":RS_rating,"52weekLow":low_52week,"52weekHigh":high_52week})
df_sp.to_csv(fname,index=False)
```
